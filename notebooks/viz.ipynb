```python
import pandas as pd
```

```python
path_to_file = '../kg-extracting-probing/data/meta/analyze_candidates.csv'
final_df = pd.read_csv(path_to_file)
```

```python
def viz(index):
    """
    enter pandas index
    """
    buffer_df = final_df.iloc[index,:]
    
    print(buffer_df.text, '\n')
    
    for unfil in eval(buffer_df.unfiltered):
        print('NC: ', unfil)
    
    print('-' * 150)

    for log in eval(buffer_df.logreg):
        print('LR: ', log)
        
    print('-' * 150)
    for com in eval(buffer_df.common_triplets):
        print('COMMON: ', com)
    
    print('-' * 150)
    for uncom in eval(buffer_df.uncommon_triplets):
        print('UNCOMMON: ', uncom)
        
    return 0
```

```python
viz(10)
```
